```
credit_data = spark.read.format('csv')\
                                .option("inferSchema", "True")\
                                .option("header", "True")\
                                .option("sep", ",")\
                                .load("dbfs:/FileStore/shared_uploads/mylokations@gmail.com/credit_train.csv")
```

```
credit_data.show(5)
```

```
credit_data.count()
```

```
credit_data_subset = credit_data.select('loan Status','Current Loan Amount','Annual Income').show(5)
```

```
credit_data.select('loan Status','Current Loan Amount','Annual Income').limit(10).show()
```

```
credit_data.filter(credit_data['Annual Income'].isNull()).show()
```

```
credit_data = credit_data.dropna()
```

```
credit_data.count()
```

```
credit_data.select('Customer ID','Annual Income','Home Ownership','Bankruptcies').where(credit_data.Bankruptcies > 0).show()
```

```
credit_data.select('Customer ID','Annual Income','Home Ownership','Tax Liens').filter((credit_data['Annual income'] >100000) & (credit_data['Tax Liens'] > 0)).show()
```

```
credit_data.select('Customer ID','Home Ownership').filter(credit_data['Home Ownership'].isin(['Home Mortgage','Rent'])).show()
```

```
credit_data.select('Customer ID','Annual Income','Monthly Debt')\
            .withColumnRenamed('Annual Income','Income')\
            .withColumnRenamed('Monthly Debt','Monthly Debt Payment')\
            .show()
```

```
credit_data.select('Customer ID','Annual Income','Monthly Debt')\
            .withColumn('Savings', credit_data['Annual Income'] -12 * credit_data['Monthly Debt'])\
            .show()
```

```
credit_data.select('Customer ID','Annual Income','Monthly Debt')\
            .orderBy(credit_data['Monthly Debt'].desc())\
            .show()
```

```
credit_data.select('Customer ID','Years of Credit History','Years in Current Job')\
            .orderBy(credit_data['Years of Credit History'].asc())\
            .show()
```

```
credit_data.groupBy('Loan Status').count().show()
```

```
credit_data.groupBy('purpose').agg({'Current Loan Amount':'sum'}).show()
```

```
credit_data.groupBy('purpose').count().show()
```

```
credit_data.groupBy('purpose').count().write.csv("dbfs:/FileStore/shared_uploads/mylokations@gmail.com/count_by_loan_purpose.csv")
```
